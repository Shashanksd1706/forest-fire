```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
df=pd.read_csv('Algerian_forest_fires_dataset.csv',header=0)
```

```python
df.head()
```

```python
df.loc[ :121,'Region']=0
df.loc[122:,'Region']=1
```

```python
df[['Region']]=df[['Region']].astype(int)
df
```

```python
df=df.dropna()
df.isnull().sum()
```

```python
df.columns
```

```python
#fix spaces in columns name
df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
# Convert columns to numeric, coercing errors to NaN
columns_to_convert = ['day', 'month', 'year', 'Temperature', 'RH', 'Ws']
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Drop rows with NaN values in the specified columns
df = df.dropna(subset=columns_to_convert)

# Convert the columns to integers after handling non-numeric values
df[columns_to_convert] = df[columns_to_convert].astype(int)

# Display the dataframe info
df.info()
```

```python
#changing the other columns to float datatype

objects=[features for features in df.columns if df[features].dtype=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
#lets save the clean dataset (we dont want to store the index so we set index=False)
df.to_csv('Algerian_forest_fires_dataset_clean.csv',index=False)
```
